# How to run or import the no-fuss ternary script into your Jupyter notebook

Start by copying and saving the Python file ``mpl_ternary.py`` (8 KB) in the same folder as your Jupyter notebook. From here you have two options, run or import the scrip.

## Running the script

Run the script using the magic command ``%run mpl_ternary.py``, this will add all the functions to the scope of your Jupyter notebook as follows

```python
# run the mpl_ternary.py script
%run mpl_ternary.py
```

```python
ax.plot(x, y, 'o', markersize=12, label='a random point')
ax.legend(fontsize=14)
```

## Importing the script

In this case, you import the script into your notebook as follows

```python
import mpl_ternary as mplt
```

and then you can access the different functions using ``mplt.`` as follows

```python
fig, ax = mplt.ternary(upper_label='$FeO_{T}$', left_label='$K_2O + Na_2O$', right_label='$MgO$')
```
